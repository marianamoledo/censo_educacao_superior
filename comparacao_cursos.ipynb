```python
import os
import pandas as pd
```

```python
pasta = r'E:\censo_educ_sup\microdados\MICRODADOS_CADASTRO_CURSOS'
arquivos = os.listdir(pasta)

for arquivo in arquivos:
    if arquivo.endswith('.CSV'):
        caminho_arquivo = os.path.join(pasta, arquivo)

        df = pd.read_csv(caminho_arquivo, nrows=1, sep=';', encoding='latin1')

        print(f'\nColunas do arquivo {arquivo}:\n{df.columns.tolist()}')
```

```python
import fileinput

for arquivo in arquivos:
    if arquivo.endswith('.CSV'):
        caminho_arquivo = os.path.join(pasta, arquivo)

        df = pd.read_csv(caminho_arquivo, sep=';' , encoding='latin1')

        quantidade_colunas = len(df.columns)
        with fileinput.input(caminho_arquivo,openhook=fileinput.hook_encoded("latin1")) as f:
            quantidade_linhas = sum(1 for line in f)


        print(f'O arquivo {arquivo} possui {quantidade_colunas} colunas e {quantidade_linhas} linhas.')
```

```python
conjuntos_colunas = {}

for arquivo in arquivos:
    if arquivo.endswith('.CSV'):
        caminho_arquivo = os.path.join(pasta, arquivo)

        df = pd.read_csv(caminho_arquivo, nrows=1, sep=';', encoding='latin1')  

        conjuntos_colunas[arquivo] = set(df.columns)

for arquivo1, colunas1 in conjuntos_colunas.items():
    for arquivo2, colunas2 in conjuntos_colunas.items():
        if arquivo1 != arquivo2:
            colunas_diferentes = colunas1.symmetric_difference(colunas2)
            if colunas_diferentes:
                print(f"Colunas diferentes entre {arquivo1} e {arquivo2}: {colunas_diferentes}")
            else:
                print(f"Os arquivos {arquivo1} e {arquivo2} tm as mesmas colunas.")
```

```python
df_arquivo_2020 = pd.read_csv(r'E:\censo_educ_sup\microdados\MICRODADOS_CADASTRO_CURSOS\MICRODADOS_CADASTRO_CURSOS_2020.CSV', sep=';', encoding='latin1')

df_arquivo_2020.rename(columns={'CO_CINE_ROTULO2': 'CO_CINE_ROTULO'}, inplace=True)
```

```python
arquivos_csv = [arquivo for arquivo in os.listdir(pasta) if arquivo.endswith('.CSV') and arquivo != 'MICRODADOS_CADASTRO_CURSOS_2020.CSV']

df_combinado = pd.DataFrame()

df_combinado = pd.concat([df_combinado, df_arquivo_2020], ignore_index=True)

for arquivo in arquivos_csv:
    caminho_arquivo = os.path.join(pasta, arquivo)
    df_temporario = pd.read_csv(caminho_arquivo, sep=';', encoding='latin1')
    df_combinado = pd.concat([df_combinado, df_temporario], ignore_index=True)

df_combinado.to_csv(r'E:\censo_educ_sup\microdados\microdados_cadastro_cursos.csv', index=False)
```
