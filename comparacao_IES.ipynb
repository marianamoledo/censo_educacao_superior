```python
import os
import pandas as pd
```

```python
pasta = r'E:\censo_educ_sup\microdados\MICRODADOS_ED_SUP_IES'
arquivos = os.listdir(pasta)

for arquivo in arquivos:
    if arquivo.endswith('.CSV'):
        caminho_arquivo = os.path.join(pasta, arquivo)

        df = pd.read_csv(caminho_arquivo, nrows=1, sep=';', encoding='latin1')  # Leitura de apenas uma linha para obter os nomes das colunas

        print(f'\nColunas do arquivo {arquivo}:\n{df.columns.tolist()}')
```

```python
for arquivo in arquivos:
    if arquivo.endswith('.CSV'):
        caminho_arquivo = os.path.join(pasta, arquivo)

        df = pd.read_csv(caminho_arquivo, sep=';' , nrows=1, encoding='latin1')  # Leitura completa do arquivo

        quantidade_colunas = len(df.columns)

        print(f'O arquivo {arquivo} possui {quantidade_colunas} colunas.')
```

```python
conjuntos_colunas = {}

for arquivo in arquivos:
    if arquivo.endswith('.CSV'):
        caminho_arquivo = os.path.join(pasta, arquivo)

        df = pd.read_csv(caminho_arquivo, nrows=1, sep=';', encoding='latin1')  

        conjuntos_colunas[arquivo] = set(df.columns)

for arquivo1, colunas1 in conjuntos_colunas.items():
    for arquivo2, colunas2 in conjuntos_colunas.items():
        if arquivo1 != arquivo2:
            colunas_diferentes = colunas1.symmetric_difference(colunas2)
            if colunas_diferentes:
                print(f"Colunas diferentes entre {arquivo1} e {arquivo2}: {colunas_diferentes}")
            else:
                print(f"Os arquivos {arquivo1} e {arquivo2} têm as mesmas colunas.")
```

```python
# Carregar o arquivo MICRODADOS_CADASTRO_IES_2009 para um DataFrame
df_IES_2009 = pd.read_csv(r'E:\censo_educ_sup\microdados\MICRODADOS_ED_SUP_IES\MICRODADOS_CADASTRO_IES_2009.CSV', sep=';', encoding='latin1')

# Mapear os nomes antigos para os novos
mapeamento_colunas = {coluna: coluna.replace('DOC_EX_', 'QT_DOC_EX_') for coluna in df_IES_2009.columns if 'DOC_EX_' in coluna}
mapeamento_colunas.update({coluna_2: coluna_2.replace('QT_DOCENTE', 'QT_DOC') for coluna_2 in df_IES_2009.columns if 'QT_DOCENTE' in coluna_2})


# Renomear as colunas no DataFrame
df_IES_2009.rename(columns=mapeamento_colunas, inplace=True)
```

```python
# Carregue o arquivo CSV original em um DataFrame
caminho_arquivo_original = r'E:\censo_educ_sup\microdados\MICRODADOS_ED_SUP_IES\MICRODADOS_CADASTRO_IES_2021.CSV'
df_IES_2021 = pd.read_csv(caminho_arquivo_original, sep=';', encoding='latin1')

# Lista de colunas para excluir
colunas_para_excluir = ['CO_PROJETO', 'CO_LOCAL_OFERTA', 'NO_LOCAL_OFERTA']

# Use o método drop para excluir as colunas
df_IES_2021 = df_IES_2021.drop(colunas_para_excluir, axis=1)

# Exiba o DataFrame após a exclusão
df_IES_2021.columns

df_IES_2021.shape
```

```python
arquivos_IES_csv = [arquivo for arquivo in os.listdir(pasta) if arquivo.endswith('.CSV') and arquivo != 'MICRODADOS_CADASTRO_IES_2021.CSV' and arquivo != 'MICRODADOS_CADASTRO_IES_2009.CSV']

df_combinado = pd.DataFrame()

df_combinado = pd.concat([df_combinado, df_IES_2009, df_IES_2021], ignore_index=True)

for arquivo in arquivos_IES_csv:
    caminho_arquivo = os.path.join(pasta, arquivo)
    df_temporario = pd.read_csv(caminho_arquivo, sep=';', encoding='latin1')
    df_combinado = pd.concat([df_combinado, df_temporario], ignore_index=True)

df_combinado.to_csv(r'E:\censo_educ_sup\microdados\microdados_IES.csv', index=False)
```
