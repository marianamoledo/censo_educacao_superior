```python
import os
import pandas as pd
```

```python
pasta = r'E:\censo_educ_sup\microdados\MICRODADOS_ED_SUP_IES'
arquivos = os.listdir(pasta)

for arquivo in arquivos:
    if arquivo.endswith('.CSV'):
        caminho_arquivo = os.path.join(pasta, arquivo)

        df = pd.read_csv(caminho_arquivo, nrows=1, sep=';', encoding='latin1')  # Leitura de apenas uma linha para obter os nomes das colunas

        print(f'\nColunas do arquivo {arquivo}:\n{df.columns.tolist()}')
```

```python
for arquivo in arquivos:
    if arquivo.endswith('.CSV'):
        caminho_arquivo = os.path.join(pasta, arquivo)

        df = pd.read_csv(caminho_arquivo, sep=';' , nrows=1, encoding='latin1')  # Leitura completa do arquivo

        quantidade_colunas = len(df.columns)

        print(f'O arquivo {arquivo} possui {quantidade_colunas} colunas.')
```

```python
conjuntos_colunas = {}

for arquivo in arquivos:
    if arquivo.endswith('.CSV'):
        caminho_arquivo = os.path.join(pasta, arquivo)

        df = pd.read_csv(caminho_arquivo, nrows=1, sep=';', encoding='latin1')  

        conjuntos_colunas[arquivo] = set(df.columns)

for arquivo1, colunas1 in conjuntos_colunas.items():
    for arquivo2, colunas2 in conjuntos_colunas.items():
        if arquivo1 != arquivo2:
            colunas_diferentes = colunas1.symmetric_difference(colunas2)
            if colunas_diferentes:
                print(f"Colunas diferentes entre {arquivo1} e {arquivo2}: {colunas_diferentes}")
            else:
                print(f"Os arquivos {arquivo1} e {arquivo2} têm as mesmas colunas.")
```
